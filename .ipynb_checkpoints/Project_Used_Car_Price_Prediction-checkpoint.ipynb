Since we have too many unique values even after preprocessing we will do some modification to the values. On a close observation we can see that broadly cars belong to ['suv', 'sedan', 'convertible', 'coupe', 'wagon', 'hatchback', 'minivan', 'van', 'cab'] categories. Either an enhanced version of it or a similar one. Ex: 'clubcab' and 'xtracab', coupe and  gcoupe etc..   

So we will write a custom function where if body type contains any of the popular type then we will use the general version else we will rename it as other

```python
def replace_body(x):
    body_type=['suv', 'sedan', 'convertible', 'coupe', 'wagon', 'hatchback', 'minivan', 'van', 'cab']
    pattern='|'.join(re.escape(substring) for substring in body_type)
    flag=re.search(pattern, x, flags=re.IGNORECASE)
    if flag:
        x=flag[0]
    else:
        x='Other'
    return x
```

```python
df['body']=df['body'].apply(lambda x: replace_body(x))
```

Now we will be performing some imputation task. So inorder to prevent data leakage, we will be splitting dataset into train and test. We will transform test set based on model trained on train data. This will remove all null values.
Our method will be to create a map of table by grouping based on 'make' and 'model' and 'condition'. So we will be imputing average value based on make and model.  Similarly for categorical column we will be imputing mode based on make nad model. If we don't find any group for the instance we will be imputing average value.

```python
# Splitting Data to prevent data leakage

from sklearn.model_selection import train_test_split

train_set, test_set=train_test_split(df, test_size=0.2, random_state=42)
```

```python
dftr=train_set.copy()
dfte=test_set.copy()
```

```python

# Custom function to impute

from sklearn.base import BaseEstimator, TransformerMixin

class RefGroupMeanImputer(BaseEstimator, TransformerMixin):
    def __init__(self, ref_columns,target_column):
        self.ref_columns = ref_columns  # Store the ref_columns as an instance variable
        self.target_column=target_column
        self.tab = None  # Initialize tab as None
        self.temp=None
        
    
    def fit(self, X, y=None):
        self.temp=X.dropna(how='any')
        self.tab=pd.DataFrame(self.temp.groupby([self.ref_columns[0], self.ref_columns[1]])[self.target_column].mean())
        return self
        
    
    def transform(self, X):
        Xcop= X.copy()  # Make a copy to leave the original DataFrame intact
        X_=Xcop[Xcop[self.target_column].isna()]
        X_[self.target_column] = X_.apply(self.impute, axis=1)
        
        Xcop.loc[X_[self.target_column].index.values, self.target_column]=X_[self.target_column].values
        return Xcop[self.target_column]

    def impute(self, row):
        i, j = row[self.ref_columns[0]], row[self.ref_columns[1]]
        if (i, j) in self.tab.index:
            return self.tab.loc[(i, j), self.target_column]
        else:
            return self.temp[self.target_column].mean()
   
        
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class RefGroupModeImputer(BaseEstimator, TransformerMixin):
    def __init__(self, ref_columns,target_column):
        self.ref_columns = ref_columns  # Store the ref_columns as an instance variable
        self.target_column=target_column
        self.tab = None  # Initialize tab as None
        self.temp=None
        self.mod=None
    
    def fit(self, X, y=None):
        self.temp=X.dropna(how='any')
        self.tab=pd.DataFrame(self.temp.groupby([self.ref_columns[0], self.ref_columns[1]])[self.target_column].apply(lambda x: x.mode().iloc[0])).reset_index()
        return self
        
    
    def transform(self, X):
        self.mod=self.temp[self.target_column].mode()
        Xcop= X.copy()  # Make a copy to leave the original DataFrame intact
        X_=Xcop[Xcop[self.target_column].isna()]
        X_[self.target_column] = X_.apply(self.impute, axis=1)
        
        Xcop.loc[X_[self.target_column].index.values, self.target_column]=X_[self.target_column].values
        return Xcop[self.target_column]

    def impute(self, row):
        i, j = row[self.ref_columns[0]], row[self.ref_columns[1]]
        if (i, j) in self.tab.index:
            return self.tab.loc[(i, j), self.target_column]
        else:
            return self.mod
```

```python
rgmi=RefGroupMeanImputer(ref_columns=['make','body'], target_column='condition')
rgmi.fit(dftr)
dftr['condition']=rgmi.transform(dftr)
dfte['condition']=rgmi.transform(dfte)
```

```python
rgmoi=RefGroupModeImputer(ref_columns=['make','body'], target_column='transmission')
rgmoi.fit(dftr)
dftr['transmission']=rgmoi.transform(dftr)
dfte['transmission']=rgmoi.transform(dfte)
```

```python
import statsmodels.api as sma
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

def vif_arr(df1):
    dfvif=pd.DataFrame()
    df1=sma.add_constant(df1)
    dfvif['variables']=df1.columns
    dfvif['VIF']= [vif(df1.values, i) for i in range(df1.shape[1])]
    return dfvif
```

```python
# Building preprocessing pipeline
# Our preprocessing mainly includes of One-Hot Encoding of categorical features and scaling them by applying 
# log-transformation and Standard scaler after it

preprocess_pipe = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),[0,1,2,5,6]),
    ('log_transform', FunctionTransformer(np.log1p, validate=False),[3,4,7,8]),
    ('standard_scale', StandardScaler(),[3,4,7,8]) 
],remainder='passthrough')



df_final=pd.DataFrame(preprocess_pipe.fit_transform(X_train),index=X_train.index)  # Sample dataframe after preprocessing
```

## 4. Model Building

```python
# Linear Regression
lin_reg = LinearRegression()



lin_reg_pipe = Pipeline([
    ('preprocess_pipe', preprocess_pipe),
    ('lin_reg',lin_reg)])

lin_reg_pipe.fit(X_train,y_train)

y_train_pred = lin_reg_pipe.predict(X_train)
y_test_pred=lin_reg_pipe.predict(X_test)

print('R2 score',metrics.r2_score(y_train,y_train_pred))
print('MAE',metrics.mean_absolute_error(y_train,y_train_pred))

print('R2 score',metrics.r2_score(y_test,y_test_pred))
print('MAE',metrics.mean_absolute_error(y_test,y_test_pred))
```

```python
print('R2 score',metrics.r2_score(np.exp(y_test),np.exp(y_test_pred)))
print('MAE',metrics.mean_absolute_error(np.exp(y_test), np.exp(y_test_pred)))
```

```python
# Ridge Regression
ridge_reg = Ridge(alpha=10)

ridge_reg_pipe =Pipeline([
    ('preprocess_pipe', preprocess_pipe),
    ('ridge_reg',ridge_reg)])

ridge_reg_pipe.fit(X_train,y_train)

y_train_pred = ridge_reg_pipe.predict(X_train)
y_test_pred=ridge_reg_pipe.predict(X_test)

print('R2 score',metrics.r2_score(y_train,y_train_pred))
print('MAE',metrics.mean_absolute_error(y_train,y_train_pred))

print('R2 score',metrics.r2_score(y_test,y_test_pred))
print('MAE',metrics.mean_absolute_error(y_test,y_test_pred))
```

```python
# Lasso Regression
laso_reg = Lasso(alpha=0.001)

laso_reg_pipe = Pipeline([
    ('preprocess_pipe', preprocess_pipe),
    ('laso_reg',laso_reg)])

laso_reg_pipe.fit(X_train,y_train)

y_train_pred = laso_reg_pipe.predict(X_train)
y_test_pred=laso_reg_pipe.predict(X_test)

print('R2 score',metrics.r2_score(y_train,y_train_pred))
print('MAE',metrics.mean_absolute_error(y_train,y_train_pred))

print('R2 score',metrics.r2_score(y_test,y_test_pred))
print('MAE',metrics.mean_absolute_error(y_test,y_test_pred))
```

## 6. Conclusion and Results Discussion

We considered used car data and performed some basic data cleaning. We dropped nearly 2000 instances(data points) based on number of null values it had. We dropped all instances that had more than 4 null values. To handle null values in transmission and condition. We grouped cars based on manufacturer and body type and then imputed average values of each group to the misisng data for the group it belongs.We used custom function for the imputation purpose. We performed other basic data preprocessing.  
Once basic preprocessing is completed, we performed rigorous EDA. We checked how target variable vehicle price is dependent on other variables. We checked how manufacturer, body type, vehicle color, condition etc.. impacts. Following is quick summary of what we found:  
* The resale value was high for luxury car brands. The resale price was in low to medium range for niche brands like Ford, Chevrolet, Toyota etc..
* Convertible, Coupe, SUV and some special editions( which were marked as other) increased price of car, whereas body type like Sedan, mini-van, hatchback, wagon were relatively less priced comparing their counterpart
* Automatic Transmission vehicles were found to be expensive than manual
* With vechicle condition deteriorating, the price dropped significantly. But for luxury cars the price was still high even if condition was average.
* Black, White, Silver, Gray, Blue and Red were popular car colours and were high in numbers. So we grouped all other colours together as Other
* For interiror people preferred black, tan, gray and beige colors.
* Some luxury cars had some unique colors which affected average price during EDA
* Price depreciated with age of vehicles, but depreciation was high for affordable cars than luxurious cars.
* Odometer reading played a keyrole in determining price. And there was some correlation between odometer and age of car which is obvious but correlation was not too high( around -0.5)

Then we checked for statistical significance of each feature. We checked for skewness, correlation etc.. Though there was slight correlation among some features after encoding and log + Standard Scaler transformation. most of correlation was removed. We checked p value and VIF for each feature. We found there was very less correlation thus confirming independence of each features.  

We built pipeline to perform categorical encoding and Scaling Transformation and model fitting. We used scikileanr's Pipeline module.  

For modelling we tried with Linear Regression, Ridge, Lasso and Random Forest Regression. 
All of the four performed quite good. But RandomForest was most suitable amongst all. So we went ahead with Random Forest for Hyperparameter Tuning.  

We got best parameter at n_estimators': 150, min_samples_split: 10, min_samples_leaf: 2, max_depth: 10
