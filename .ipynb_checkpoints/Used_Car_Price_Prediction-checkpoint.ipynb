
#### Strategy to Handle null-values  

Null Values: ['make', 'model', 'trim', 'body', 'transmission', 'condition','odometer', 'color', 'interior']    


[1] If there are more than 3 or 4 null values; we will drop that particular data  

[2] After previous step we are left with same columns except 'make'. The number of null values in model, trim, color and interior are less than 1000. So we can drop these values since they contribute 1/545 = 0.1% of the total data.  
--Perform Train test split--  

[3] So now we are left with three features having null values: i.body ii. transmission iii.condition    

[4]. For condition we will be using simple-imputer mean based on model, year and make features

[5]. For model and transmission we use simple-imputer mode based on make and model

[6].We can try with KNN imputer but it works well with numerical features only.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class RefGroupMeanImputer(BaseEstimator, TransformerMixin):
    def __init__(self, ref_columns,target_column):
        self.ref_columns = ref_columns  # Store the ref_columns as an instance variable
        self.target_column=target_column
        self.tab = None  # Initialize tab as None
    
    def fit(self, X, y=None):
        temp=X.dropna(how='any')
        self.tab=pd.DataFrame(temp.groupby([self.ref_columns[0], self.ref_columns[1]])[self.target_column].mean())
        return self
        
    
    def transform(self, X):
        Xcop= X.copy()  # Make a copy to leave the original DataFrame intact
        X_=Xcop[Xcop[self.target_column].isna()]
        X_[self.target_column] = X_.apply(self.impute, axis=1)
        
        Xcop.loc[X_[self.target_column].index.values, self.target_column]=X_[self.target_column].values
        return Xcop[self.target_column]

    def impute(self, row):
        i, j = row[self.ref_columns[0]], row[self.ref_columns[1]]
        if (i, j) in self.tab.index:
            return self.tab.loc[(i, j), self.target_column]
        else:
            return self.tab[self.target_column].mean()
   
        
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class RefGroupModeImputer(BaseEstimator, TransformerMixin):
    def __init__(self, ref_columns,target_column):
        self.ref_columns = ref_columns  # Store the ref_columns as an instance variable
        self.target_column=target_column
        self.tab = None  # Initialize tab as None
    
    def fit(self, X, y=None):
        temp=X.dropna(how='any')
        self.tab=pd.DataFrame(temp.groupby([self.ref_columns[0], self.ref_columns[1]])[self.target_column].apply(lambda x: x.mode().iloc[0])).reset_index()
        return self
        
    
    def transform(self, X):
        Xcop= X.copy()  # Make a copy to leave the original DataFrame intact
        X_=Xcop[Xcop[self.target_column].isna()]
        X_[self.target_column] = X_.apply(self.impute, axis=1)
        
        Xcop.loc[X_[self.target_column].index.values, self.target_column]=X_[self.target_column].values
        return Xcop[self.target_column]

    def impute(self, row):
        i, j = row[self.ref_columns[0]], row[self.ref_columns[1]]
        if (i, j) in self.tab.index:
            return self.tab.loc[(i, j), self.target_column]
        else:
            return self.tab[self.target_column].mode()
   
```

#### VIF

```python
import pandas as pd
import numpy as np
import statsmodels.api as sma
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
```

```python
def vif_arr(df1):
    dfvif=pd.DataFrame()
    df1=sma.add_constant(df1)
    dfvif['variables']=df1.columns
    dfvif['VIF']= [vif(df1.values, i) for i in range(df1.shape[1])]
    return dfvif
    
```

```python
vif_arr(dftr[['odometer', 'mmr','condition','year_difference']])
```

##### Please note that in the python code I have added a column of intercept/constant to my data set before calculating the VIFs. This is because the variance_inflation_factor function in python does not assume the intercept by default while calculating the VIFs. Hence, often we may come across very different results in R and Python output. For details, please see this discussion 

### p-Value

```python
import statsmodels.api as sm

# X should contain your independent variables (features)
X = dftr[['odometer', 'mmr','condition','year_difference']]  # Replace with your actual feature columns

# Add a constant term to the independent variables (for intercept)
X = sm.add_constant(X)

# y is your dependent variable
y = dftr['sellingprice']  # Replace with your actual target variable

# Fit a linear regression model
model = sm.OLS(y, X).fit()

# Get the summary of the regression model
summary = model.summary()

# Find the p-values for each feature's coefficient
p_values = model.pvalues

# Print the summary to view the p-values and other statistics
print(summary)

# Access individual p-values for each feature
print("P-values for each feature:")
print(p_values)
```
